DATA TRANSFORMATIONS

```python
# CONVERTING CATEGORICAL VALUES TO NUMERICAL VALUES

train_data = train_data.replace(to_replace="low", value=0)
train_data = train_data.replace(to_replace="medium", value=0.5)
train_data = train_data.replace(to_replace="high", value=1)

train_data.head()
```

```python
# REMOVING THE NECESSARY COLUMNS

train_data = train_data.drop(['emp_id'], axis=1)

train_data.head()
```

```python
# CONVERTING THE COLUMNS WHICH ARE CATEGORICAL DATA TYPE INTO A NUMERIC TYPE FOR TRAINING THE MODEL

from sklearn.preprocessing import LabelEncoder
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)


train_data['department'] = LabelEncoder().fit_transform(train_data['department'])

train_data.head()
```

```python
# SPLITTING THE DATASET AS BASED ON DEPENDENT AND INDEPENDENT VARIABLES [COLUMNS/ATTRIBUTES]

X = train_data.drop('left', axis=1) # INDEPENDENT VARIABLES
Y = train_data['left'] # DEPENDENT VARIABLES

print("ORIGINAL SHAPE OF THE DATASET : ", train_data.shape)
print("SHAPE OF DATA CONTAINING INDEPENDENT VARIBALES : ", X.shape)
print("SHAPE OF DATA CONTAINING DEPENDENT VARIBALES : ", Y.shape)
```

```python
# SPLITTING THE DATASET INTO TRAINING AND TESTING BY SPECIFYING THE SIZE

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# CREATING A LOGISTIC REGRESSION MODEL
log_reg = LogisticRegression(max_iter=10000)

# FITTING THE MODEL TO THE TRAINING DATA
log_reg.fit(X_train, Y_train)

# PREDICTING THE MODEL TO THE TESTING DATA
y_pred = log_reg.predict(X_test)

# FINDING THE ACCURACY OF THE MODEL
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy   :", round(accuracy, 2) * 100)
```

```python
from sklearn.ensemble import RandomForestClassifier


# CREATING A RANDOM FOREST CLASSIFIER MODEL
rf_classifier = RandomForestClassifier()

# FITTING THE MODEL TO THE TRAINING DATA
rf_classifier.fit(X_train, Y_train)

# PREDICTING THE MODEL TO THE TESTING DATA
y_pred = rf_classifier.predict(X_test)

# FINDING THE ACCURACY OF THE MODEL
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy   :", round(accuracy, 2) * 100)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# CREATING A GRADIENT BOOSTER CLASSIFIER MODEL
gb_classifier = GradientBoostingClassifier()

# FITTING THE MODEL TO THE TRAINING DATA
gb_classifier.fit(X_train, Y_train)

# PREDICTING THE MODEL TO THE TESTING DATA
y_pred = gb_classifier.predict(X_test)

# FINDING THE ACCURACY OF THE MODEL
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy   :", round(accuracy, 2) * 100)
```

```python
from sklearn.svm import SVC

# CREATING A SUPPORT VECTOR CLASSIFIER MODEL
svm_classifier = SVC()

# FITTING THE MODEL TO THE TRAINING DATA
svm_classifier.fit(X_train, Y_train)

# PREDICTING THE MODEL TO THE TESTING DATA
y_pred = svm_classifier.predict(X_test)

# FINDING THE ACCURACY OF THE MODEL
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy   :", round(accuracy, 2) * 100)
```

```python
import xgboost as xgb

# CREATING A XG BOOST CLASSIFIER MODEL
xgb_classifier = xgb.XGBClassifier()

# FITTING THE MODEL TO THE TRAINING DATA
xgb_classifier.fit(X_train, Y_train)

# PREDICTING THE MODEL TO THE TESTING DATA
y_pred = xgb_classifier.predict(X_test)

# FINDING THE ACCURACY OF THE MODEL
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy   :", round(accuracy, 2) * 100)
```
